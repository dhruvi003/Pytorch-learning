```python
"""
Autograd
- it is difficult to find derivative of nested functions
- the whole neural network is dependent on derivations

- core component that provides automatic differentiation of tensor operations.
this enables gradient computation which is essential for training ml model

"""
```

```python
import torch
```

```python
x = torch.tensor(3.0,requires_grad=True)
# requires_grad -- requires_gradient, so pytorch will understand that we wish to calculate derivative of x.
# so, pytorch will track the operations that we will be performing on x
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
y.backward()
# pytorch will give derivative
```

```python
x.grad
```
