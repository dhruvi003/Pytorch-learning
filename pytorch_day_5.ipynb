```python
# create model class
import torch
import torch.nn as nn
```

```python
# nn
class Model(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        # we will create our layer here
        self.linear = nn.Linear(num_features, 1) # input, output
        self.activation = nn.Sigmoid()

    def forward(self, features):
        # output calculation
        out = self.linear(features)
        out = self.activation(out) #y_pred
        return out
```

```python
# dataset
features = torch.rand(10,5)
```

```python
model = Model(features.shape[1])
```

```python
## creating nn with hidden layer
class NNTwo(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.layer1 = nn.Linear(num_features,3)
        self.layer2 = nn.Linear(3,1)
        self.activation1 = nn.ReLU()
        self.activation2 = nn.Sigmoid()


    def forward(self, features):
        out = self.layer1(features)
        out = self.activation1(out)
        out = self.layer2(out)
        out = self.activation2(out)
        return out
```

```python
features = torch.rand(10,5)
```

```python
features
```

```python
model = NNTwo(features.shape[1])
```

```python
# simplify forward function - we use sequential container
class Test(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features, 3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
    )

    def forward(self, features):
        return self.network(features)
```
