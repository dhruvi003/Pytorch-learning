```python
# let's see difference of time between cpu and gpu
import time

size = 10000

mtrx_1 = torch.randn(size, size)
mtrx_2 = torch.randn(size, size)

start_time = time.time()
result_cpu = torch.matmul(mtrx_1, mtrx_2)
cpu_end_time = time.time() - start_time

print("CPU TIME:", cpu_end_time)

start_time = time.time()
result_gpu = torch.matmul(mtrx_1.to('cuda'), mtrx_2.to('cuda'))
torch.cuda.synchronize()
gpu_end_time = time.time() - start_time

print("GPU TIME:", gpu_end_time)
```
